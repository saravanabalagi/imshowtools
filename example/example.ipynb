```python
from imshowtools import imshow, cvshow
import tensorflow as tf
import numpy as np
import cv2
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
# x_train, x_test = x_train / 255.0, x_test / 255.0
[x_train.shape, y_train.shape], [x_test.shape, y_test.shape]
```

## Quick Plot

```python
imshow(x_train[0])
imshow(x_train[0], x_train[1], x_train[2])
```

```python
imshow(*x_train[:20], cmap='binary')
imshow(*x_train[:100], cmap='binary', size=(10, 10))
```

## Plot and Obtain Images

```python
image_1 = imshow(*x_train[:100], cmap='binary', return_image=True)
image_2 = imshow(*x_train[:100], cmap='binary', return_image="RGBA")
image_3 = imshow(*x_train[:100], cmap='binary', return_image="RGB")
image_4 = imshow(*x_train[:100], cmap='binary', return_image="BW")
print(image_1.shape)
print(image_2.shape)
print(image_3.shape)
print(image_4.shape)

# cv2.imwrite("saved_sample.png", image_1)
# do stuff with 'image_1' or even
imshow(image_1, image_2, image_3, image_4)
```

## Rows and Columns

```python
imshow(*x_train[:15], cmap='Purples', rows=1)
imshow(*x_train[:24], cmap='Greens', columns=4)
```

## OpenCV Images

```python
lenna = cv2.imread('lenna.png')
imshow(lenna)
cvshow(lenna)
imshow(lenna, mode='BGR')

image = imshow(*[lenna for _ in range(12)], return_image="BW")
print(image.shape)
imshow(image)
```

## Note 
All images (including single channel images) are shown as images, not colormaps. If you need to show them as colormaps, explicitly use cmap argument. You can discern the difference from the below code.

```python
image = np.full([28, 28, 1], fill_value=1.0)
image2 = np.full([28, 28, 1], fill_value=0.0)
image3 = np.full([28, 28, 1], fill_value=0.5)
```

```python
print('imshowtools')
imshow(image, image2, image3)
```

```python
print('matplotlib')
imshow(image, image2, image3, cmap='viridis')
```

```python
print('matplotlib')
imshow(image, image2, image3, cmap='gray')
```
